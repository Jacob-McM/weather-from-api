```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime

import scipy.stats as st

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities_data.csv"
```

```python
# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)
```

```python
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count
len(cities)
```

```python
cities
```

```python
#Desired url = https://api.openweathermap.org/data/2.5/weather?q={city name},{units}&appid={API key}
url = "https://api.openweathermap.org/data/2.5/weather?"
units = "&units=imperial"
query_url = (f"{url}&appid={weather_api_key}&q=")
```

```python

# for city in cities:
#     counter = counter + 1
#     print(counter)
#     time.sleep(60/50)
```

```python
#Lists for append
lat = []
long = []
date = []
country = [] 
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
weather_request = []
cities_found = []
```

```python
counter = 0
for city in cities:
    try:
        counter = counter + 1
        weather_report = requests.get(f"{query_url}{city},{units}").json()        
        print(f"Trying to get data for City {counter} of {len(cities)} : {city}")
        weather_request.append(weather_report)
        lat.append(weather_report["coord"]["lat"])
        long.append(weather_report["coord"]["lon"])
        date.append(weather_report["dt"])
        country.append(weather_report["sys"]["country"])
        max_temp.append(weather_report["main"]["temp_max"])
        humidity.append(weather_report["main"]["humidity"])
        cloudiness.append(weather_report["clouds"]["all"])
        wind_speed.append(weather_report["wind"]["speed"])
        cities_found.append(weather_report["name"])
        time.sleep(60/50)
    except (RuntimeError, TypeError, NameError, KeyError):
        print(f"Skipped city {counter}, {city}. Something wasn't found. Maybe the city itself?")   
```

```python
fixed_dates= []
for n in date:
    fix_date = datetime.fromtimestamp(n).strftime("%x %X")
    fixed_dates.append(fix_date)
```

```python
weather_df = pd.DataFrame({
    "Country":country,
    "City": cities_found,
    "Latitude":lat,
    "Long": long,
    "Temperature(F)": max_temp,
    "Humidity %": humidity,
    "Cloudiness %": cloudiness,
    "Wind Speed(mph)": wind_speed,
    "Date": fixed_dates
    
})
weather_df
```

# Linear Regression

 In our circumstances, r<sup>2</sup> values range from a minimum of 0 to a maximum of 1. 0 is the absolute absence of a linear relationship, with 1 being a certainty of the presence of a linear relationship. 

```python
norhtern_hemi = weather_df.loc[weather_df['Latitude'] >= 0].copy()
southern_hemi = weather_df.loc[weather_df['Latitude'] < 0].copy()
print(f'The amount of cities in the Northern Hemisphere is: {norhtern_hemi.index.size}')
print(f'The amount of cities in the Southern Hemisphere is: {southern_hemi.index.size}')
```

## Maximum Temperature vs Latitude 

### Northern Hemisphere

```python
x_values = norhtern_hemi['Latitude']
y_values = norhtern_hemi['Temperature(F)']

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values,y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(10,5),facecolor='white')
plt.rc('font', size=12)
plt.grid(visible=True)
plt.scatter(x_values,y_values,marker="o", facecolors='aliceblue', edgecolors="black",s=100,linewidths=1.1)
plt.plot(x_values,regress_values,"r-",linewidth=3,alpha=.8)
plt.annotate(line_eq,(18,36),fontsize=15,color="black")
plt.title("Maximum Temperature vs Latitude in the Northern Hemisphere")
plt.suptitle(weather_df['Date'][0],style='italic')
plt.ylabel("Max Temperature")
plt.xlabel("Latitude")
plt.grid(visible=True)
plt.savefig("../Images/northern_temp_lat_reg.png")
print(f"The r2 value is: {rvalue**2}")
```

The idea of this linear regression plot is to see the linear relationship between an areas maximum temperature and its place in the northern hemisphere. At an r<sup>2</sup> .799, my analysis would be that there is a linear relationship between the maximum temperature of a location and it's place in the norhtern hemipshere. If your location is above the equator, based on the cities used in this analysis, there is a relationship between your place in the world and the temperature you experience. Important to note, based on the amount of datapoints, the r<sup>2</sup> value will lean towards being larger than a smaller dataset.

### Southern Hemisphere

```python
x_values = southern_hemi['Latitude']
y_values = southern_hemi['Temperature(F)']

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values,y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(10,5),facecolor='white')
plt.rc('font', size=12)
plt.grid(visible=True)
plt.scatter(x_values,y_values,marker="o", facecolors='aliceblue', edgecolors="black",s=100,linewidths=1.1)
plt.plot(x_values,regress_values,"r-",linewidth=3,alpha=.8)
plt.annotate(line_eq,(-18,36),fontsize=15,color="r",)
plt.title("Maximum Temperature vs Latitude in the Southern Hemisphere")
plt.suptitle(weather_df['Date'][0],style='italic')
plt.ylabel("Max Temperature")
plt.xlabel("Latitude")
plt.grid(visible=True)
plt.savefig("../Images/southern_temp_lat_reg.png")
print(f"The r2 value is: {rvalue**2}")
```

With an r<sup>2</sup> value of .352 it is likely that there is not a strong linear relationship between locality in the southern hemipshere and the temperature experienced. To note, the amount of data points for the southern hemisphere in this analysis are smaller than the northern hemisphere, which does affect the scale of their r values. Another note, is that there is far more habitated landmass in the Norhtern Hemisphere, and the extremes are less likely to be habitated.
